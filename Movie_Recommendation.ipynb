```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies=movies.merge(credits,on='title')
movies.head(1)
```

```python
#We will make a  list of which columns are important for making our model
# genres
# id
# keywords
# title
# overview
# cast
# crew
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
# check any missing data
movies.isnull().sum()
```

```python
# as there are 3 missing data in overview we will remove those 3 rows
movies=movies.dropna()
movies.isnull().sum()
```

```python
# to check is there any duplicate data present and then removing it
movies.duplicated().sum()
#no duplicate row present
```

```python
movies.iloc[0].genres
#here the genre is in string form we need to convert it into array list form
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#creating a function to extract name keyword value from dictionaries of genres in movies dataframe
#here we will use ast module to convert string lsit in array list with the inbuild function litarel_eval()
def convert(obj):
  List=[]
  for i in ast.literal_eval(obj):
    List.append(i["name"])
  return List
```

```python
#now applying the above function convert to genre column of movies
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
  List=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter !=3:
      List.append(i["name"])
      counter+=1
    else:
      break
  return List
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
#here we will fetch the director from the movies crew for that function is created
def fetch_director(obj):
  List=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      List.append(i["name"])
  return List
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
#now all the column are preprocessed
movies.head()
```

```python
#now we will convert string overview into list of string
movies['overview']= movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#now we will remove blank spaces between two words eg.actor name Sam Worthington replace by SamWorthington
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies.head()
```

```python
#now concatinate all the four column to single tags column
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
#now create newdataframe
new_df=movies[['movie_id','title','tags']]
new_df
```

```python
#now join all tags list into the string
new_df['tags']= new_df['tags'].apply(lambda x:" ".join(x))
new_df.head()
```

```python
#here now we will apply steming which means words like[action,actioned,actions]==>[action,action,action]
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
#here we will vectorize all the tags into vector number
#we will use scikit learn library for this purpose
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
#now we will calculate cosine distance i.e angle between the two vectors
#beacuse dealing with high dimensional vector we should avoid euclidian distance
#and make use of cosine angle distance where small angle means similar movies and large angle shows disimilar movies
from sklearn.metrics.pairwise import cosine_similarity
#here similarity near 1 means highly similar movie and similarity near 0 means dissimilar movie
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity.shape
#it will calcuate smilarity between all movie for each movie
```

```python
def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df['title'][i[0]])
```

```python
recommend('Avatar')
```
